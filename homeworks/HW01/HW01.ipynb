# 2. Примеры создания массивов:

### Создание массивов numpy из списков

```python
lst1 = [1, 3, 3, 5, 2, 4, 8, 8]
lst2 = [[1, 3, 3, 5], [2, 4, 8, 8]]
```

### Создание массивов с помощью функций

```python
'''
zeros() - создает матрицу из нулей (n x m)
'''
z_arr = np.zeros((2,3))
```

```python
'''
eye() - создает матрицу из нулей (n x m) с единицами на заданной k-той диагонали
'''
eye_arr = np.eye(4, k=1, dtype=int)
```

```python
'''
random() - создает матрицу (n x m) со случайными числами от 0 до 1 
'''
rng_arr = random((5,5)).round(2)
```

```python
'''
arange(start,stop,step) - создает матрицу с числами от start до stop с шагом step
'''
range_arr = np.arange(1,10,2)
```

```python
'''
linspace() - создает массив (последовательность) равномерно распределенных чисел в заданном интервале (start, end, num)
'''
lin_arr = np.linspace(1, 10, 100)
```

# 5. Примеры векторных операций и базовых статистик

### поэлементное сложение или умножение массивов

```python
A = np.array([1, 2, 3, 4, 5])
B = np.array([1, 4, 4, 7, 7])
```

```python
A * B # Умножение поэлементное
```

```python
np.dot(A, B) # Скалярное произведение
```

```python
C = np.array([[1,2,4,5,4],[1,7,7,7,7]])
D = np.array([[1,2],[4,5],[4,1],[7,7],[7,7]])
```

```python
np.matmul(C,D) # Умножение матриц
```

```python
A + B # Поэлементное сложение
```

```python
A - B # Поэлементная разница
```

### вычисление средней величины (mean) или другой простой статистики

```python
mu, sigma = 2, 10
lin2 = np.random.normal(mu, sigma, size=100).round(2) # генерирует случайные выборки из нормального (гауссовского) распределения.
```

```python
import matplotlib.pyplot as plt
plt.hist(lin2, 30, density=True)
plt.show()
```

```python
lin2
```

```python
np.var(lin2).round(2) # Дисперсия
```

```python
np.std(lin2).round(2) # Станд. отклонение
```

```python
np.mean(lin2).round(2) # Арифм. среднее
```
